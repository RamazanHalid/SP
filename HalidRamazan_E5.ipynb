#Exercise 9.5: Singular values

```python
from scipy.linalg import toeplitz
import numpy
import random
import scipy
n = int(input('input n: '))

C = []
p = float(input('input P: '))
for i in range(0, n):
    tmp = []
    for j in range(0, n):
        if random.uniform(0, 1) > p:
            tmp.append(0)
        else:
            tmp.append(1)
    C.append(tmp)

U, sigma, VT = scipy.linalg.svd(C)
print(sigma)
```

#Exercise 10.1: Least squares

```python

import numpy as np
import scipy.optimize as opt
 
m, n = 100, 50
A = np.random.rand(m, n)  
b = np.random.random(m)
 
# Now find x = argminx k Ax − bk 2.
res = opt.lsq_linear(A, b)
x = res.x
 
# The residual's norm should be printed.
residual = b - np.dot(A, x)
norm = np.linalg.norm(residual) #Frobenius norm
 
print(norm)
```

# Exercise 10.2: Optimization

```python
import numpy as np
import scipy.optimize as opt 
# f(x) = sin 2(x − 2)e −x 2
def f(x):  
	res = np.power(np.sin(x-2), 2) * np.exp(-1*(x**2))
	# Change this function to the polar opposite.
	return -1 * res
    
# Find the function's greatest value.
res = opt.minimize_scalar(f)
x = res.x
# Finding the minimum of the opposing function is the same as finding the minimum of the opposite function.
print(-1*f(x))
```

#Exercise 10.3: Pairwise distances

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import scipy.spatial.distance as dis
  
n, m = 50, 100
X = np.random.rand(n, m)
de = dis.pdist(X)
 
# Consider the following scenario: there are n cities, and we are given their coordinates in two columns.
x = []
y = []
for a in range(0, n//2):
	x = x + [a for _ in range(1, n)]
	y = y + [b for b in range(1, n)]
 
x = np.array(x)
y = np.array(y)
 
# Now we need a good table that informs us how far away each pair of cities is.
fig = plt.figure()
ax = Axes3D(fig)
 
# creat a 3-D table
plt.title("Distance between two cities")  
ax.set_xlabel ('Firt City')
ax.set_ylabel('Second City')
ax.set_zlabel('Distance')
 
ax.scatter(x, y, de, c='r')
plt.show()
```

#Exercise 11.1: Plotting a function

```python
import matplotlib.pyplot as plt
import numpy as np


size = 100
f, ax = plt.subplots()
ones = np.ones(size)
x = np.linspace(0., 2., size)
y1 = 0.5 * (ones - np.cos(2 * (x - ones * 2)))
y2 = np.power(np.e, -1 * np.multiply(x, x))
y = np.multiply(y1, y2)
ax.plot(x, y)
ax.set_xlabel("this is x label")
ax.set_ylabel("this is y label")
ax.set_title("$sin^2(x-2){e^{-x^2}}$")
plt.tight_layout()
plt.show()
```

#Exercise 11.2: Data

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
X = np.random.random_sample((20, 10)) * 10
b = np.random.random(10) * 3 - 1.5
z = np.random.normal((20,))
y = X.dot(b) + z
b_ = np.array(np.linalg.lstsq(X, y, rcond=-1)[0])
x = np.arange(0, 10)

f, ax = plt.subplots()
ax.set_xlim(0, 9)
ax.set_ylim(-1.5, 1.5)
ax.set_xlabel("index")
ax.set_ylabel("value")
ax.plot(x, b, 'rx', label='True coefficients')
ax.plot(x, b_, 'bo', label='Estimated coefficients')
plt.hlines(0, 0, 9, colors='k', linestyle="--")
plt.tight_layout()
plt.show()
```
