#Exercise 9.4: Power iteration

```python
import time
n=200
def PowerIter(Z,esp):
    my_array = np.ones(n)
    my_array = my_array.reshape(n,1)
    max_my_array = np.max(my_array)
    sub = np.inf
    iteration = 0
    start_time = time.clock()
    while sub > esp:
        my_array = Z@my_array
        temp = np.max(my_array)
        sub = np.fabs(temp-max_my_array)
        max_my_array = temp
        my_array = my_array / max_my_array
        iteration += 1
    end_time = time.clock()
    return max_my_array, iteration, end_time-start_time, my_array

esp = 1e-5
Z = np.random.normal(0,1,(n,n))
print("Maxeig, iterations, time/s, eigVec")
print(PowerIter(Z,esp))
```
